```python
import pandas as pd

df = pd.read_csv("/content/archive (1).zip")

df.columns = [
    "Customer_ID",
    "City",
    "State",
    "Service_History",
    "Common_Problem",
    "Solution_Used",
    "Vehicle_Company"
]

df.head()
```

```python
df['Vehicle_Company'].value_counts()
```

```python
df['Common_Problem'].value_counts()
```

```python
df['Solution_Used'].value_counts()
```

```python
df['City'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Common_Problem'].value_counts().plot(kind='bar')
plt.title("Most Common Vehicle Problems")
plt.xlabel("Problem Type")
plt.ylabel("Frequency")
plt.xticks(rotation=90)
plt.show()
```

```python
df['Solution_Used'].value_counts().plot(kind='bar')
plt.title("Most Used Repair Solutions")
plt.xticks(rotation=90)
plt.show()
```

```python
df['Vehicle_Company'].value_counts().plot(kind='bar')
plt.title("Service Requests by Vehicle Company")
plt.show()
```

```python
df['City'].value_counts().head(10).plot(kind='bar')
plt.title("Top Cities with Most Services")
plt.show()
```

```python
df.to_csv("clean_vehicle_service_data.csv", index=False)
```
